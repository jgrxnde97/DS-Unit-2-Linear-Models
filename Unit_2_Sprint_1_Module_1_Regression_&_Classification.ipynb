```python
import pandas as pd
df = pd.read_csv(DATA_PATH+'apartments/renthop-nyc.csv')

print(df.shape)
df.head()
```

```python
# Calculating mean/difference of bathrooms
guess = df['bedrooms'].mean()
errors = guess - df['bedrooms']
mean_abs_error = errors.abs().mean()
print(mean_abs_error)
```

```python
# Plotting trend of bedrooms and bathrooms
import plotly.express as px
px.scatter(df, x='bedrooms', y='bathrooms', trendline='ols')
```

```python
df.describe()
```

```python
# Improting appropriate estimator from Scikit-Learn
from sklearn.linear_model import LinearRegression

# Instantiate class
model = LinearRegression()

# Arrange X features matrix and y target vector
features = ['bedrooms']
target = 'bathrooms'
X_train = df[features]
y_train = df[target]

# Fitting the model
model.fit(X_train, y_train)
```

```python
# Apply model to new data
bedrooms = 9
X_test = [[bedrooms]]

model.predict([[bedrooms]]) # do not add quotations
```

```python
y_pred = model.predict(X_test)
print(f'Predicted bedrooms {X_test} for bathrooms:' + str(y_pred))
```

```python
model
```

```python
model.coef_
```

```python
# Cartesian plotting formula  y = mx + b
m = model.coef_[0]
b = model.intercept_

print('y = mx + b')
print(f'y = {m:.0f}*x + {b:.0f}')
print(f'bathroom in apt. = {m:.0f}*bedrooms + {b:.0f}')
```

```python
# Applying model to new data and explaining coefficient
def predict (bedrooms):
  y_pred = model.predict([[bedrooms]])
  estimate = y_pred[0]
  coefficient = model.coef_[0]
  result = f'${estimate:,.0f} beds per bath {bedrooms:,.0f} in'
  explanation = f'In this linear regression; for each additional bedroom there is {coefficient}'
  return result + explanation

predict(10)
```
